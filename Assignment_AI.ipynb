```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
%matplotlib inline
```

## Sigmoid Activation Function

It is defined as:
$$
\sigma(z)=\frac {1}{1+e^{-z}}
$$
The plot is as below.

```python
pi=np.pi
z =np.linspace(-4*pi,4*pi,1000)
def f(z):
    return 1/(1+np.exp(-z))
plt.plot(z, f(z))
plt. xlabel('$z$', fontsize=14)
plt.ylabel('$\sigma (z)$', fontsize=14)
plt.title('Sigmoid Activation Function', fontsize=14)
plt.legend('plot')
plt.grid(True)
plt.show()
```

```python
z=sp.Symbol('z')
def g(z):
    return 1/(1+sp.exp(-z))
sp.pprint(sp.diff(g(z),z))
```

```python
a =np.linspace(-4*pi,4*pi,1000)
def f(a):
    return (np.exp(-a))/((1+np.exp(-a))**2)
plt.plot(a, f(a))
plt. xlabel('$z$', fontsize=14)
plt.ylabel("$\sigma' (z)$", fontsize=14)
plt.title(' Derivative of Sigmoid Activation Function', fontsize=14)
plt.legend('plot')
plt.grid(True)
plt.show()
```
